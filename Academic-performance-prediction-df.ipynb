```python
# Perform one-hot encoding
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df_encoded = academic_perform_df.apply(le.fit_transform)
print(df_encoded)
```

```python
X = df_encoded.drop(columns = ['G3'])
y = academic_perform_df[['G3']]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

# **TASK#3 CREATE TRAINING AND TESTING DATASET**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X
```

# **TASK#4 TRAIN A LINEAR REGRESSION MODEL IN SCIKIT-LEARN**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

regresssion_model_sklearn = LinearRegression(fit_intercept = True)
regresssion_model_sklearn.fit(X_train, y_train)
```

```python
regresssion_model_sklearn_accuracy = regresssion_model_sklearn.score(X_test, y_test)
regresssion_model_sklearn_accuracy
```

```python
predictions = regresssion_model_sklearn.predict(X_test)  #List of all predictions
for x in range(len(predictions)):
    print(np.round(predictions[x]),y_test[x])
```

```python
print('Linear Model Coefficient (m): ', regresssion_model_sklearn.coef_)
print('Linear Model Coefficient (b): ', regresssion_model_sklearn.intercept_)
```

# **TASK#5 EVALUATE TRAINED MODEL PERFORMANCE**

```python
# Make prediction
y_predict = regresssion_model_sklearn.predict(X_test)
y_predict
```

```python
# Plot the scaled result
plt.plot(y_test, y_predict, "^", color = 'b')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```
